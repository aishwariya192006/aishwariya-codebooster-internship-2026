```python
import os
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
os.environ["GROQ_API_KEY"]="Groq_Api_Key"
llm=ChatGroq(model="llama-3.3-70b-versatile",temperature=0)
prompt=PromptTemplate(
    input_variables=["instruction"],
    template="Follow the instruction and provide the output clearly:\n\nInstruction:{instruction}\n\nAnswer"

)
parser=StrOutputParser()
chain=prompt| llm | parser
response=chain.invoke({
    "instruction":"Translate 'Thank,bye' to german"
})
print("\ngerman:\n",response)
```

```python
import os
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
os.environ["GROQ_API_KEY"]="Groq_api_key"
llm=ChatGroq(model="llama-3.3-70b-versatile",temperature=0)
prompt=PromptTemplate(
    input_variables=["instruction"],
    template=(
        """
        english:coffee and tea
        german:kaffee und tee
        english:cookies or milk
        german:kekse oder milch
        Follow the instruction below:\n\nInstruction:{instruction}\n\nAnswer"
        answer:
        """
)
)
parser=StrOutputParser()
chain=prompt| llm | parser
response=chain.invoke({
    "instruction":"Translate 'coffee or tee' to german"
})
print("\ngerman:\n",response)
```
