```python
import os
from langchain_groq  import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

groq_api=input("Enter your Groq API Key: ").strip()
os.environ["GROQ_API_KEY"] = groq_api

llm=ChatGroq(model="llama-3.3-70b-versatile",temperature=0)
#no shot

prompt=PromptTemplate(
    input_variables=["instruction"],
    template="""You are a professional LinkedIn content writer and personal branding expert.

Your task is to generate a high-quality LinkedIn post from the user's instruction,emoji,token_size.

Follow these rules:

1. Understand the user's instruction, achievement, experience, opinion, project, event, learning, or announcement.
2. Create an attention-grabbing opening hook and based on the temp value increase or decrease the creativity level of the output.
3. Write in a professional yet natural and human tone.
4. The length of the paragraph should be of the length of the token_size and for one line allow only 10 tokens in output.
5. Highlight key insights, lessons learned, challenges overcome, or impact created.
6. Maintain a positive and engaging style.
7. Avoid generic AI phrases, clichГ©s, and excessive corporate jargon.
8. Include relevant emojis only if the user give 'yes' to the input variable emoji otherwise don't include emojis.
9. End with a question or call-to-action to encourage engagement.
10. Add 3вЂ“8 relevant hashtags.
11. Do not explain your reasoning.
12. Return only the final LinkedIn post.

Input:
{instruction}
{emoji}
{token_size}

Answer:"""
)

parser = StrOutputParser()

chain = prompt | llm | parser


instruct=input("Enter the instruction: ")
token_size=input("Enter the token size: ")
emoji=input("Do you want to include emojis? (yes/no): ")
temp=input("Enter the creativity/temperature level on the basis from 1 to 0 for high to low: ")

response = chain.invoke({
    "instruction":instruct,
    "token_size":token_size,
    "emoji":emoji,
    "temperature":temp
})

print(response)
```
