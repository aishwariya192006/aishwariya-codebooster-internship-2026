````python
import os
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

os.environ["GROQ_API_KEY"]="YOUR_GROQ_API_KEY"

llm= ChatGroq(model="llama-3.3-70b-versatile",temperature=0)

prompt=PromptTemplate(
    input_variables=["instruction"],
    template="Follow the instruction and provide the clearly:\n\nInstruction:{instruction}\n\nAnswer:"
)
parser = StrOutputParser()
chain = prompt | llm | parser
response = chain.invoke({
    "instruction":"Translate 'Happy morning , how was your day?' to Tamil"
})
print("\n The given instruction is done by zero-shot output",response)
````

````python
import os
from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser

os.environ["GROQ_API_KEY"] = "YOUR_GROQ_API_KEY"
llm = ChatGroq(model = "llama-3.3-70b-versatile", temperature=0)
prompt = PromptTemplate(
    input_variables = ["instruction"],
    template = "I eat rice. – நான் சாதம் சாப்பிடுகிறேன். She is my friend. – அவள் என் தோழி.This is a pen. – இது ஒரு பேனா.We play cricket. – நாங்கள் கிரிக்கெட் விளையாடுகிறோம்.The cat is sleeping. – பூனை தூங்கிக்கொண்டிருக்கிறது. \n\n Now, follow the instruction below: Instruction:{instruction}\n\nAnswer:"
)

parser = StrOutputParser()
chain = prompt | llm | parser
responce = chain.invoke({
    "instruction":"Translate 'Where are you?' to tamil"
})

print("\n The given instruction is done by few-shot output : \n", responce)
````
