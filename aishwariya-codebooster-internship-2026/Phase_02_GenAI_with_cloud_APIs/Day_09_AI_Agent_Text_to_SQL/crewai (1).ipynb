```python
import os
os.environ["GROQ_API_KEY"]="groq_api_key"
```

```python
from crewai import Agent,Task,Crew

def res():
  researcher=Agent(
      role='research analyst',
      goal='Identify top ai trends for 2025',
      backstory='Experienced technology researcher with 10+ years of industry experience',
      verbose=True,
      llm='groq/llama-3.70b-8192'

  )
  research_task=Task(
      description='List and analyze the top 3 AI trends for 2025 with examples',
      expected='comphrensive report with detailed explanation for explanation for each trend',
      agent=researcher
  )
  crew=Crew(
      agents=[researcher],
      tasks=[research_task],
      verbose=True
  )
  return crew.kickoff()
result=res()
print(result)
```
